```python
passages = [
    "Machine learning enables computers to learn from data without explicit programming.",
    "Deep learning uses neural networks with many hidden layers to model complex patterns.",
    "Transformers are attention-based neural networks used in NLP tasks.",
    "Sentence Transformers generate embeddings for semantic similarity search.",
    "Chroma is a vector database for storing and querying embeddings.",
    "Vector databases support fast similarity search for AI applications.",
    "Natural language processing helps machines understand human language.",
    "Large language models are trained on massive text corpora."
]

print("Number of passages:", len(passages))
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(passages, convert_to_numpy=True)

print("Embedding shape:", embeddings.shape)
```

```python
import chromadb
from chromadb.config import Settings

client = chromadb.Client(
    Settings(anonymized_telemetry=False)
)

collection = client.create_collection(name="text_embeddings")
```

```python
collection.add(
    documents=passages,
    embeddings=embeddings.tolist(),
    ids=[str(i) for i in range(len(passages))]
)

print("Documents added to Chroma ✅")
```

```python
query = "What is a vector database used for?"
query_embedding = model.encode([query], convert_to_numpy=True)

results = collection.query(
    query_embeddings=query_embedding.tolist(),
    n_results=3
)

print("Query:", query)
print("\nTop matching passages:")
for doc in results["documents"][0]:
    print("-", doc)
```
